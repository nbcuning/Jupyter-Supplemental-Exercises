## [1.1] What is an API?


Imagine you’re hit with a sudden craving for a burger. You open Foodpanda and search for “burger” near you. The app doesn’t already store all restaurant menus on your phone or device. Instead, a fast digital request-response process begins:

- Step 1: **API Client:** When you search for “burger” on Foodpanda, the app (API client) starts the process by sending a request through an API asking which restaurants near you offer burger options.

- Step 2: **API Request:** The app sends an API request containing the search term and your location. This request is forwarded through the API system so it can be processed by the server.

- Step 3: **API Server:** The API forwards this request to Foodpanda’s backend servers, which connect to multiple restaurant databases and retrieve the relevant information.

- Step 4: **API Response:** The server returns data in a structured format, usually JSON. This includes restaurant names, menu items, prices, availability, and distance from your location. The API then sends this response back to the app, which formats it into a readable list of available burger options for the user.

![](food_panda.png)

**Note:** Applications on their own would quickly become outdated if they had to store all their information internally. Instead, they rely on external systems to fetch data whenever it is needed. This allows the app to remain lightweight and efficient while still providing constantly updated information.

**APIs** are formally defined as a set of rules that enable different software applications to communicate and exchange data or functionality in a structured manner. They allow developers to integrate services from external systems without having to build them from scratch. APIs expose only the necessary data required for a request while keeping internal system details hidden, ensuring both security and efficiency. A common format used for transmitting this data is **JSON.**

## [1.2] JSON as a Language in APIs

**JSON,** or JavaScript Object Notation, is a lightweight, text-based data-interchange format. Although it is derived from a subset of the JavaScript programming language, it is entirely language-independent. This means that a system written in Python can easily generate JSON that is then consumed by a system written in Java, Ruby, or C#.

The reason why JSON is widely used in the industry is that:

- JSON is **human readable** because it is written in plain text with a simple and clear structure, making it easy for developers to understand and debug.
- JSON is **machine usable** because it follows a strict, standardized format that computers can quickly parse, process, and exchange between systems.

Other alternative formats include:
- XML (eXtensible Markup Language)
- CSV (Comma Separated Values)
- YAML

In Python, the process of converting a data structure into a format suitable for transmission is called Encoding (or Serialization). The reverse process—converting a string back into a programmatic object—is called Decoding (or Deserialization).

Python’s built-in json library facilitates these operations through two primary functions:

| Function    | Action | Instruction |
|---------|-----|--------|
| json.dumps()   | Encode  | Python Object $\rightarrow$ JSON String  |
| json.loads()     | Decode  | JSON String $\rightarrow$ Python Object   |

## [1.3] The Structure and Syntax of JSON

JSON is built on two primary structures that mirror common programming data types:

- **Objects:** A collection of name/value pairs. In JSON, an object begins with a left curly brace `{` and ends with a right curly brace `}.` Each name is followed by a colon :, and pairs are separated by commas.
- **Arrays:** An ordered list of values. An array begins with a left bracket `[` and ends with a right bracket `].`

JSON is a strictly typed format. Every value in a JSON structure must belong to one of the six categories defined by the standard.

| Data Type | Description | Example |
|----------|----------|----------|
| Number    | Represents numeric values, including integers and floating-point decimals. No quotes used.     | `"id": 42`     |
| String    | A sequence of Unicode characters used for text. Must be enclosed in double quotes.     | `"title": "Back in Black"`     |
| Boolean    | A logical value representing truth or falsity. Must be lowercase true or false.     | `"is_classic": true`     |
| Null    | A special value indicating that a key has no associated data or is intentionally empty.     | `"producer": null`     |
| Array    | An ordered list of values enclosed in square brackets `[ ].` Can contain various types.     | `"tracks": [ ... ]`     |
| Object    | A set of key-value pairs enclosed in curly braces `{ }.` Often used for nesting data.     | `{"id": 1, "title": "Hells Bells"}`     |

## [1.4] Example: How JSON Transfers Structured Data

**JSON** is widely used as the standard format for web APIs because it can **represent structured data,** including information in which items are nested within other items. In this example, we simulate the process of sending and receiving a user profile.

We define our dataset as a Python dictionary. In this case, the structure is more complex because it includes nested data. The `location` field is itself an object (another dictionary) containing multiple related attributes, while the `interests` field is a list of strings that represents multiple values under a single category. This combination of nesting and lists allows the dataset to represent real-world information more accurately.


```python
import json

# Creating the dataset (Python Dictionary)
user_profile = {
    "user_id": 1024,
    "username": "the_encoder",
    "is_active": True,
    "bio": None,
    "location": {
        "city": "Quezon City",
        "timezone": "PHT"
    },
    "interests": ["Machine Learning", "Python", "Data Science"]
}

print("Dataset defined successfully.")
```

To transfer this profile over a network, we must convert the Python object into a JSON string. We use `json.dumps()` for the transformation along with `indent=4` for readability.

```python
# Encoding the Python object into a JSON formatted string
json_encode = json.dumps(user_profile, indent=4)

print("Transferred Data (JSON String):")
print(json_encode)
```

Now that the profile has been converted to JSON, we can see how the data are translated into the JSON standard data types:

| Data | Python Representation | JSON Representation |
|----------|----------|----------|
| User ID   | `int` | `Number` |
| Status   | `True` | `true` |
| Bio   | `None` | `null` |
| Location   | `dict` | `Object` |
| Interests   | `list` | `Array` |

Once the data reaches its destination, the receiving system decodes the string back into an object using `json.loads()`:

```python
# Decoding the JSON string back into a Python dictionary
received_profile = json.loads(json_encode)

# Accessing nested structured data
user_city = received_profile['location']['city']
primary_interest = received_profile['interests'][0]

print(f"Decoded Data:")
print(f"User: {received_profile['username']}")
print(f"Location: {user_city}")
print(f"Main Interest: {primary_interest}")
```

This shows that JSON is not only for simple lists. It can also **store structured data** with multiple levels, such as a user’s settings, location, and preferences, all in one text format. This makes it easy to send data between different programs and systems, even if they use **different programming languages.**
